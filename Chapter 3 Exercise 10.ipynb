```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

##3.10 a-d

```python
carseats = pd.read_csv('data/carseats.csv')
model = smf.ols('Sales ~ Price + Urban + US', data=carseats)
fit = model.fit()
print(fit.summary())
```

##3.10 e-g

```python
model2 = smf.ols('Sales ~ Price + US', data=carseats)
fit2 = model2.fit()
print(fit2.summary())
```

##3.10 h

```python
plt.figure(figsize=(18, 6))
ax = plt.subplot(131)
plt.plot(carseats['Price'], fit2.resid, 'o', [20, 200], [0, 0], 'r-')
ax = plt.subplot(132)
fig = sm.qqplot(fit2.resid, line='s', ax=ax)
ax = plt.subplot(133)
fig = sm.graphics.influence_plot(fit2, ax=ax, size=10)
xmin, xmax = plt.xlim(-0.001, 0.04)
```
